```python
def opencalls(df, stock_list):
  open_calls = []  # Assuming open_call is a list you want to append to
  for stock in stock_list:
    # Order the DataFrame by 'Stock' and 'Date'
        new_df = df[df['Stock'] == stock].sort_values(by=['Stock', 'Date'], ascending=[True, False])
        date_list = sorted(new_df.Date.unique(), reverse = True)
        subset_df = new_df[(new_df['Date'] == date_list[0]) & (~new_df['Reco'].isin(["Exit", "EXIT"]))]
        if not subset_df.empty:
           open_calls.append(subset_df.iloc[0])
        else:
          continue

  return open_calls
```
