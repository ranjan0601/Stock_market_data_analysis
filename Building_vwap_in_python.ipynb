```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
data = yf.Ticker("HDFCBANK.BO")
```

```python
data1 = data.history(interval = '15m', period = '60d')
```

```python
data1
```

```python
data1['VWAP'] = (data1['Volume'] * data1['Close']).cumsum()/data1['Volume'].cumsum()
```

```python
data1
```

```python
last_row = data1.iloc[-1]
secound_last_row = data1.iloc[-2]

print(f"last_row['Close'] : {last_row['Close']}")
print(f"last_row['VWAP'] : {last_row['VWAP']}")
print(f"secound_last_row['Close'] : {secound_last_row['Close']}")
print(f"secound_last_row['VWAP'] : {secound_last_row['VWAP']}")
```

```python
if secound_last_row['Close'] > secound_last_row['VWAP'] and last_row['Close'] < last_row['VWAP']:
  print("Cross Under")
if secound_last_row['Close'] < secound_last_row['VWAP'] and last_row['Close'] > last_row['VWAP']:
  print("Cross Over")
else:
  print('No Cross')
```

```python
from dateutil.relativedelta import relativedelta
```

```python
One_month_ago =  current_date - relativedelta(months=1)
Three_month_ago = current_date - relativedelta(months=3)
Six_month_ago = current_date - relativedelta(months=6)
print(current_date)
print(One_month_ago)
print(Three_month_ago)
print(Six_month_ago)
```

```python
date_list = [current_date, One_month_ago, Three_month_ago, Six_month_ago]
date_list
```

```python
ava_data = []
# prev_date = [-1,-2,-3,-4,-5,-6,-7,-8]
for elm in date_list:
  i = -1
  if elm in (angel_data['Date'].values):
    ava_data.append(elm)
  else:
    while True:
      offset_date = elm + timedelta(days = i)
      if offset_date in (angel_data['Date'].values):
        ava_data.append(offset_date)
        break
      i -= 1
```

```python
def average_price(df, ava_data):
    ava_data = sorted(ava_data, reverse = True)
    result_data = []
    start_date = ava_data[0]
    for i in range(len(ava_data)-1):  # Loop over the date ranges
        # start_date = ava_data[i]
        end_date = ava_data[i + 1]

        # Filter the DataFrame based on the date range
        month_data = df[(df['Date'] <= start_date) & (df['Date'] >= end_date)]

        # Calculate and store the average price for the month
        if not month_data.empty:
            average_price = round(month_data['Close'].mean(),2)
            result_data.append({'Start Date': start_date, 'End Date': end_date, 'Average Price': average_price})

    return result_data

# Example usage:
result = average_price(angel_data, ava_data)
# print(result)
result
```
